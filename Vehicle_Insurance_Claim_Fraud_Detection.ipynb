#### Now we are going to label encode the categorical value

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
for c in categorical:
    encoder.fit(data[c])
    data[c]=encoder.transform(data[c])
    mapping=dict(zip(encoder.classes_,encoder.transform(encoder.classes_)))
    print(c,':',mapping)
```

#### Here age is the only continuous value so we have to scale it

```python
sns.kdeplot(x=data['Age'])
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
```

```python
from sklearn.model_selection import train_test_split
x=data.drop(['FraudFound','PolicyNumber'],axis=1)
x['Age']=scale.fit_transform(x[['Age']])
y=data['FraudFound']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

#### Since we are having 30+ columns it makes ML ineffective and slow so we are going to use SelectKBest

```python
from sklearn.feature_selection import SelectKBest, chi2, f_classif
fs1=SelectKBest(f_classif,k = 10)
fs1.fit(x_train , y_train)
```

```python
fs1.get_support()
```

```python
x_train=x_train[x_train.columns[fs1.get_support()]]
x_test=x_test[x_test.columns[fs1.get_support()]]
```

```python
x_train.head(10)
```

```python
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
rfc=RandomForestClassifier()
rfc.fit(x_train, y_train)
res=rfc.predict(x_test)
score=accuracy_score(y_test,res)
print(score*100)
```

```python
clas_report=classification_report(y_test,res)
print(clas_report)
```

```python
matrix = confusion_matrix(y_test,res)
print(matrix)
```

```python
sns.heatmap(matrix,annot=True,fmt='0.1f')
plt.show()
```

#### In the above  steps we can concur that there is imbalance of data.We can see that zero is predicted as zero in many places whereas one prediction is really not up to the mark bcause there is an imabalance.


#### To deal whith this we are going to do" ---Under Sampling--- one of the methods for imbalance data. "


```python
yes=data[data['FraudFound']==1]
```

```python
no=data[data['FraudFound']==0]
```

```python
print(yes.shape,no.shape)
```

```python
no_new=no.sample(892)
```

```python
print(yes.shape,no_new.shape)
```

```python
df=pd.concat([yes,no_new],axis=0)
```

```python
df.shape
```

```python
df.columns
```

### Here we are going to split new df and apply the algorithm

```python
x_new=df.drop(['FraudFound','PolicyNumber'],axis=1)
y_new=df.FraudFound
```

```python
x_train2,x_test2,y_train2,y_test2 = train_test_split(x_new,y_new,test_size=0.2,random_state=42)
```

```python
print(x_train2.shape,y_train2.shape)
print(x_test2.shape,y_test2.shape)
```

```python
rfc_n=RandomForestClassifier(n_estimators=205)
rfc_n.fit(x_train2,y_train2)
res_n=rfc_n.predict(x_test2)
score_rfc=accuracy_score(y_test2,res_n)*100
print(score_rfc)
```

```python
mat=confusion_matrix(y_test2,res_n)
sns.heatmap(mat,annot=True,fmt='0.2f')
plt.show()
```

```python
params_xg = {'n_estimators': [100, 200, 500,750],'learning_rate': [0.01,0.05,0.1,0.15],'booster': ['gbtree', 'gblinear']
            }
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import GridSearchCV
xgb = GridSearchCV(XGBClassifier(), params_xg,scoring='roc_auc')
xgb.fit(x_train2, y_train2)
```

```python
ans_2=xgb.predict(x_test2)
score_xgb=accuracy_score(y_test2,ans_2)*100
print(score_xgb)
```

```python
mat2=confusion_matrix(y_test2,ans_2)
sns.heatmap(mat2,annot=True,fmt='.1f')
plt.show()
```

```python
param_svc={'kernel':['rbf','linear'],'gamma':[0.05,0.1,0.5,1],'degree':[3,4,5],'C':[1.0,1.5,2.0]}
```

```python
from sklearn.svm import SVC
svc_n=GridSearchCV(SVC(),param_svc,refit=True,verbose=3)
svc_n.fit(x_train2,y_train2)
ans_3=svc_n.predict(x_test2)
```

```python
score_svc=accuracy_score(y_test2,ans_3)*100
print(score_svc)
```

```python
mat3=confusion_matrix(y_test2,ans_3)
sns.heatmap(mat3,annot=True,fmt='.2f')
```
